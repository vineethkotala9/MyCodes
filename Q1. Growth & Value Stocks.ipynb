```python
df2=df1.loc[['T','INTC','META','GOOGL','AAPL','ADBE','TSLA'],['trailingPE']]
df2
```

```python
### Trailing PE of <18 is considered as value and viceversa
```

```python
df4=df2.loc[df2['trailingPE']<18]
df4
```
