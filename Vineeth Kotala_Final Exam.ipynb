```python
def feasible_n(n):
    np.random.seed(0)
    a2=np.random.rand(n,2)
    df2=pd.DataFrame(data={'return':np.zeros(n),'std':np.zeros(n)})
    for i in range(n):
        a2[i]/=sum(a2[i])
        df2['return'][i]=np.dot(df1.mean(),a2[i])*250
        df2['std'][i]=np.dot(a2[i].T, np.dot(df1.cov()*250, a2[i]))**0.5
    return df2
```

```python
def amortization(r,n,loan):
    payment=npf.pmt(rate=r,nper=n,pv=-loan)
    df=pd.DataFrame(data={'balance':np.zeros(n),'interest':np.zeros(n),'principal':np.zeros(n)})
    for i in range(n):
        if i==0:
            df['balance'][i]=loan
            df['interest'][i]=df['balance'][i]*r
            df['principal'][i]=payment-df['interest'][i]
        else:
            df['balance'][i]= df['balance'][i-1]-df['principal'][i-1]
            df['interest'][i]=df['balance'][i]*r
            df['principal'][i]=payment-df['interest'][i]
    return round(df,2)
```

```python
amortization(0.05,20,250000)
```
