```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

```python
### 1st Question
```

```python
df1 = pdr.get_data_yahoo('GOOGL', start='2020-10-01',end='2020-10-30')
df1
```

```python
df1['Simple return']=df1['Adj Close'].pct_change()
df1
```

```python
df1['Cum return'] = df1['Simple return'].add(1).cumprod().sub(1)
df1
```

```python
df1['Log return'] = np.log(df1['Adj Close'] / df1['Adj Close'].shift(1))
df1
```
